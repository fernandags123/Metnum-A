```python
import pandas as pd
import matplotlib.pyplot as plt


def hitung_galat(nilai_sebenarnya, nilai_hampiran):
    galat_absolut = abs(nilai_sebenarnya - nilai_hampiran)
    galat_relatif = galat_absolut / abs(nilai_sebenarnya) if nilai_sebenarnya != 0 else None
    galat_persen = galat_relatif * 100 if galat_relatif is not None else None
    return galat_absolut, galat_relatif, galat_persen

nilai_sebenarnya = float(input("Masukkan nilai sebenarnya: "))
nilai_hampiran_list = [float(x) for x in input("Masukkan daftar nilai hampiran: ").split()]


data = []
for nh in nilai_hampiran_list:
    abs_err, rel_err, pct_err = hitung_galat(nilai_sebenarnya, nh)
    data.append([nh, abs_err, rel_err, pct_err])

df = pd.DataFrame(data, columns=["Nilai Hampiran", "Galat Absolut", "Galat Relatif", "Galat Persen (%)"])


print("\n HASIL PERHITUNGAN GALAT ")
display(df.style.format({
    "Galat Absolut": "{:.2f}",
    "Galat Relatif": "{:.2f}",
    "Galat Persen (%)": "{:.2f}"
}).set_table_styles([
    {"selector": "th", "props": [("background-color", "#4CAF50"), ("color", "white")]}
]))


plt.figure(figsize=(8,5))
plt.plot(df["Nilai Hampiran"], df["Galat Absolut"], marker='o', label="Galat Absolut")
plt.plot(df["Nilai Hampiran"], df["Galat Persen (%)"], marker='s', label="Galat Persen (%)")
plt.axhline(0, color="gray", linestyle="--")
plt.xlabel("Nilai Hampiran")
plt.ylabel("Galat")
plt.title("Grafik Galat vs Nilai Hampiran")
plt.legend()
plt.grid(True)
plt.show()
```
